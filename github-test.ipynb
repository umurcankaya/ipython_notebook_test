```python
t = np.linspace(0,15,1500)
x1 = np.sin(10*t)
x2 = np.sin(11*t)
```

```python
plt.plot(t, x1, '-g', alpha=0.5, label='x1')
plt.plot(t, x2, '-b', alpha=0.5, label='x2')
plt.legend(loc='best')
```
